```python
import numpy as np
import pandas as pd
```

```python
## Upload the movies csv dataset with no titles
movies = pd.read_csv('movie_rating_dataset_no_titles.csv', header=None)
print(movies.head())
```

```python
# Add column headers
movies.columns = ['ID','Title','Release_Date', 'Overview', 'Popularity']
print(movies.head())
```

```python
#Upload the Ratings dataset with no titles
ratings = pd.read_csv('movie_ratings_dataset_2_no_titles.csv', header=None)
print(ratings.head())
```

```python
# Add column headers
ratings.columns = ['ID', 'Vote_Average', 'Vote_Count', 'Video']
print(ratings.head())
```

```python
# Merge the datasets into one dataset
data = pd.merge(movies, ratings, on =['ID','ID'])
print(data.head())
```

```python
# Look at the info for the data
data.info()
```

```python
# Look for missing values
data.isnull().sum()
```

```python
# Pull the rows with missing values in the Overview column
data[data.isna().any(axis=1)]
```

```python
# Take a look at the distribution of the ratings
import plotly.express as px
fig = px.histogram(data, 'Vote_Average')
fig.show()
```
